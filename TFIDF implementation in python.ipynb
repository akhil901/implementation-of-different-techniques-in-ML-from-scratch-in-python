# TFIDF implementation from scratch in python

```python
from collections import Counter
import pandas as pd
def fit(text):
    total_features=[]
    for i in text:
        total_features+=(i.split()) #total number of words in the sentence including stop words 
    tokenize=[]
    for i in text:
        tokenize.append(i.split()) #each sentance is tokenized 
    count_vec=[]  
    for i in tokenize:
        count_vec.append(Counter(i)) #number of times each word occuring in the sentance
    features=list(set(total_features)) #total number of unique features
    features1=dict.fromkeys(features,0)#initalizing a dict with 0 for every feature
    return features1,count_vec,tokenize,total_features
```

```python
text = [ "this is  random sentence",
       "you can write yourown sentence",
       "this is TFIDF implementation",
       "this is the last sentence"]
for i in range(len(text)):
    text[i]=text[i].lower()
features1,count_vec,x,y = fit(text)
```

```python
import math
N=len(count_vec)+1
def transform(count_vec,features1,total):
    
    for i in count_vec:
        c = dict(i)
        values = c.values()
        length = sum(values)
        for key in features1.keys():
            if key  in c.keys():
                if not isinstance(features1[key], list):
                    features1[key]=[features1[key]]
                val = c[key]
                tf=float(val/length) #calculating term frequency
                cnt=1
                for list_ in total:
                    if key in list_:
                        cnt+=1
                idf = 1+(math.log((N/cnt))) #inverse document frequency the sklearns formula for idf is
                tf_idf = tf*idf             #different so to tally with the ans used the sklearns formula
                features1[key].append(tf_idf)
            else:
                if not isinstance(features1[key], list):
                    features1[key]=[features1[key]]
                features1[key].append(0)
    return features1 
```

```python
tfidf = transform(count_vec,features1,x)
df = pd.DataFrame(tfidf)
df =df.loc[1:].sort_index(axis =1)
df
```

```python
cols = df.columns
from sklearn.preprocessing import Normalizer
norm = Normalizer().fit_transform(df)
```

```python
tfidf_vec = pd.DataFrame(norm,columns=cols)
tfidf_vec
```
